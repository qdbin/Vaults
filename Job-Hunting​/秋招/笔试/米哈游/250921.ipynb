### 02-打击序列计数

```python
import sys 

n=int(input())

for line in sys.stdin:
    s = line.split()[0]
    cur_count,pre_count,cur_c,pre_c,res=0,0,'','',[]

    # 计算字符，example："LLRRRRLLRR"==>[2,4,2,2]
    for i in range(len(s)):
        cur_c=c=s[i]
        if i==0:
            pre_c=c
            pre_count=1
            continue
        elif i==len(s)-1:
            if cur_c==pre_c:
                res.append(pre_count+1)
            else:
                res.append(pre_count)
                res.append(1)
            break
        else:
            if cur_c==pre_c:
                cur_count=pre_count+1
                cur_c=c
                pre_c=cur_c
                pre_count=cur_count
            else:
                cur_count=1
                res.append(pre_count)
                cur_c=c
                pre_c=cur_c
                pre_count=cur_count
                continue
    print(f'res的结果是：{res}')

    # 爬楼梯实现
    dp=[1]*(len(s)+1)
    dp[2]=2
    for i in range(3,len(s)+1):
        dp[i]=dp[i-1]+dp[i-2]
    print(f'dp的结果是：{dp}')

    # 最终结果
    ans=0
    for x in res:
        ans+=dp[x]
        
    print(ans)
```
