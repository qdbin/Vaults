```python
a,b=map(int,input().split())
arr=list(map(int,input().split()))
# a,b=5,2
# arr=[1,2,2,3,1]

res=sum(set(arr))

dic={}      # value:count
for i in range(len(arr)):
    if str(arr[i]) in dic:
        dic[str(arr[i])]=dic[str(arr[i])]+1
    else:
        dic[str(arr[i])]=1

l=[]    # (value,count)
for key in dic:
    if dic[key]>1:
        l.append((int(key),dic[key]))
# l.sort()
# l.reverse()


for x,y in l:
    if b-y<=0:
        res+=x*(b-1)
        break
    else:
        b=b-y
        res+=x*(y-1)

print(res)
```
