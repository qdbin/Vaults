## Numpy ndarray对象
> Numpy定义一个n维数组对象，简称ndarray对象，它是一系列相同类型元素组成的数组集合

> ndarry对象采用数组的索引机制，将数组中每个元素映射到内存上，并按照一定的布局对内存块排列，常用布局方式有两种，即按行或按列。数组中每个元素占有大小相同的内存块。

**语法格式:**
```python
"""
参数 描述说明
object 表示一个数组序列。
dtype 可选参数，通过它可以更改数组的数据类型。
copy 可选参数，表示数组能否被复制，默认True。
order 内存布局创建数组方式：C(行序列)/F(列序列)/A(默认)。
ndim 用于指定数组的维度。
"""
numpy.array(object, dtype = None, copy = True, order = None,ndmin = 0)
```

````python
# 创建一维数组
import numpy as np
a=np.array([1,2,3])
print(a)

# 创建多维数组
b=np.array([[1,2,3],[4,5,6]])
print(b)

# 改变数组元素的数据类型
c=np.array([1,2,3],dtype=np.float16)
d=np.array([1,2,3],dtype="complex")

# ndim查看数组维数
arr = np.array([[1,2,3],[4,5,6]])
print(arr.ndim)

# reshape数组变维
a=np.array([[1,2],[3,4],[5,6]])
print(a)
print(a.shape)  # out:(3,2)
b=a.reshape(2,3)
print(b)
print(b.shape)  # out:(2,3)
````
