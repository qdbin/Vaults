```python
import sys

def main():
    data = sys.stdin.read().split()
    if not data:
        return

    n,m,q=tuple(map(int,input().split()))
    arr = list(map(int, input().split()))
    x_list = list(map(int, input().split()))
    
    arr_mod = [a % m for a in arr]
    a, b, c = arr_mod
    perms = [
        (a, b, c),
        (a, c, b),
        (b, a, c),
        (b, c, a),
        (c, a, b),
        (c, b, a)
    ]
    
    results = []
    for x in x_list:
        x_mod = x % m
        x2_mod = (x_mod * x_mod) % m
        max_val = 0
        for coef in perms:
            a_p, b_p, c_p = coef
            value = (a_p * x2_mod) % m
            value = (value + (b_p * x_mod)) % m
            value = (value + c_p) % m
            if value > max_val:
                max_val = value
        results.append(str(max_val))
        
    print(" ".join(results))

if __name__ == "__main__":
    main()
```
