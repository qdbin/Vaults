```python
# 导入相关的包
import matplotlib.pyplot as plt

# 国庆期间南京最低气温
date_lists = ["10/01","10/02","10/03","10/04","10/05","10/06","10/07"] # x轴
temp_min_lists = [16,14,13,15,17,14,15] # y轴


""" plot()绘制x轴、y轴数据，并设置展示格式
- 模板：plt.plot(x, y, "格式控制字符串", 关键字=参数)
- 格式控制： r:红色    -.：虚线    o：原点
"""
plt.plot(date_lists, temp_min_lists,'r-.o')
plt.show()
```

### Matplotlib.pyplot的接口汇总
> `pyplot`模块提供绘图的各种函数，比如创建一个画布，在画布中创建一个绘图区域，或是在绘图区域添加一些线、标签、设置线条形状和颜色等
#### 绘图函数
| 函数名称 | 描述 |
| --------  | ----             |
| bar       | 条形图            |
| barh      |水平条形图         |
| boxplot   | 箱型图            |
| hist      | 直方图            |
| his2d     | 2D直方图          |
| pie       | 饼状图            |
| plot      | 在坐标轴画线或标记 |
| scatter   | 散点图            |
| polar     | 极坐标图          |

#### Axis函数
| 函数名称      | 描述              |
| ------------- | ----------------- |
| axes          | 在画布中添加轴    |
| text          | 向轴添加文本      |
| title         | 设置标题          |
| xlabel/ylabel | 设置x/y轴标签     |
| xlim/ylim     | 设置x/y轴区间大小 |
| xcale/yscale  | 设置x/y轴缩放比例 |
| xticks/yticks | 设置x/y轴刻标     |


#### Image函数

| 函数名称 | 描述           |
| --------  | --------------                  |
| imread    | 从文件中读取图像的数据并形成数组   |
| imsave    | 将数组另存为图像文件              |
| imshow    | 在数轴区域内显示图像              |

#### Figure函数

| 函数名称 | 描述           |
| -------- | -------------- |
| figtext  | 在画布添加文本 |
| figure   | 创建一个新画布 |
| show     | 显示           |
| savefig  | 保存当前画布   |
| close    | 关闭画布窗口   |
